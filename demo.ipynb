```python
import requests
import time

def get_jobs_from_indeed():
    url = "https://indeed12.p.rapidapi.com/jobs/search"
    querystring = {"query": "Information technology", "sort": "date"}

    headers = {
        "x-rapidapi-key": "e46546b67emsh1da2cc270e90508p17c823jsnc5bc19b57230",
        "x-rapidapi-host": "indeed12.p.rapidapi.com"
    }

    page = 1  # Bắt đầu từ trang 1
    while True:
        querystring["page_id"] = page  # Thêm tham số trang vào query
        response = requests.get(url, headers=headers, params=querystring)
        
        if response.status_code == 200:
            data = response.json()
            
            # Kiểm tra nếu có công việc trong phản hồi
            if data.get('count', 0) > 0:
                for job in data.get('hits', []):
                    print(f"Job Title: {job.get('title')}")
                    print(f"Location: {job.get('location')}")
                    print(f"Salary: {job.get('salary', 'Not available')}")
                    print("----------")
            
            # Kiểm tra nếu còn trang tiếp theo
            next_page_id = data.get('next_page_id')
            if next_page_id:
                page += 1  # Chuyển sang trang tiếp theo
                print(f"Chuyển sang trang {page}...")
            else:
                print("Không còn trang tiếp theo.")
                break  # Kết thúc vòng lặp nếu không có trang tiếp theo
        elif response.status_code == 429:
            # Lấy thời gian chờ từ header "Retry-After"
            retry_after = response.headers.get('Retry-After')
            if retry_after:
                print(f"Lỗi 429: Quá nhiều yêu cầu. Đang chờ {retry_after} giây để thử lại...")
                time.sleep(int(retry_after))  # Chờ theo thời gian được chỉ định trong "Retry-After"
            else:
                print("Lỗi 429: Quá nhiều yêu cầu. Đang chờ 30 giây để thử lại...")
                time.sleep(30)  # Nếu không có "Retry-After", chờ 30 giây
        else:
            print(f"Lỗi khi lấy dữ liệu: {response.status_code}")
            break

# Gọi hàm để lấy công việc
get_jobs_from_indeed()
```

```python
import requests
import pandas as pd

# 1. Khởi tạo thông tin API
url = "https://jobs-api14.p.rapidapi.com/v2/list"
headers = {
    "x-rapidapi-key": "e46546b67emsh1da2cc270e90508p17c823jsnc5bc19b57230",
    "x-rapidapi-host": "jobs-api14.p.rapidapi.com"
}

jobs_list = []
page = 1
max_pages = 20  # Giới hạn số trang tối đa

while page <= max_pages:  # Giới hạn số trang
    querystring = {"query": "Web Developer", "location": "United States", "page": page}
    response = requests.get(url, headers=headers, params=querystring)
    
    if response.status_code == 200:
        data = response.json()
        jobs_data = data.get("jobs", [])
        
        # Dừng nếu không còn dữ liệu
        if not jobs_data:
            print("Đã tải xong tất cả dữ liệu.")
            break
        
        for job in jobs_data:
            jobs_list.append({
                "Title": job.get("title", "Not Available"),
                "Company": job.get("company", "Not Available"),
                "Location": job.get("location", "Not Available"),
                "Description": job.get("description", "Not Available"),
                "Posted Date": job.get("date", "Not Available")
            })
        
        print(f"Đã tải xong trang {page}.")
        page += 1
    else:
        print(f"Lỗi khi tải dữ liệu từ API: {response.status_code}")
        break

# 4. Tạo DataFrame từ dữ liệu đã thu thập
df = pd.DataFrame(jobs_list)

# 5. Hiển thị DataFrame
print(df)

# 6. Lưu vào file CSV
df.to_csv("jobs_data_limited_pages.csv", index=False, encoding="utf-8")
```

```python
import re
def extract_12_digits(description):
    # Tìm kiếm 12 số ngay sau dấu $
    match = re.search(r'\$(\d{12})', description)
    
    if match:
        return match.group(1)  # Trả về 12 số sau dấu $
    return None  # Nếu không tìm thấy 12 số sau $

# Áp dụng hàm trích xuất 12 số
df['Salary'] = df['Description'].apply(extract_12_digits)
df
```
