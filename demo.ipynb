```python
import requests
import csv

def get_jobs_from_indeed():
    url = "https://indeed12.p.rapidapi.com/jobs/search"
    querystring = {"query": "Information technology", "sort": "date"}

    headers = {
        "x-rapidapi-key": "e46546b67emsh1da2cc270e90508p17c823jsnc5bc19b57230",
        "x-rapidapi-host": "indeed12.p.rapidapi.com"
    }

    with open("jobs_data.csv", mode="w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)
        # Ghi tiêu đề cột
        writer.writerow(["Job Title", "Location", "Salary"])
        
        page = 1 
        while True:
            querystring["page_id"] = page 
            response = requests.get(url, headers=headers, params=querystring)
            
            if response.status_code == 200:
                data = response.json()
                
    
                if data.get('count', 0) > 0:
                    for job in data.get('hits', []):
                        title = job.get('title', 'Not available')
                        location = job.get('location', 'Not available')
                        salary = job.get('salary', 'Not available')
                        
                        
                        writer.writerow([title, location, salary])
                        
                        
                        print(f"Job Title: {title}")
                        print(f"Location: {location}")
                        print(f"Salary: {salary}")
                        print("----------")
                
            
                next_page_id = data.get('next_page_id')
                if next_page_id:
                    page += 1  
                    print(f"Chuyển sang trang {page}...")
                else:
                    print("Không còn trang tiếp theo.")
                    break  
            else:
                print(f"Lỗi khi lấy dữ liệu: {response.status_code}")
                break


get_jobs_from_indeed()
```

```python
import requests
import pandas as pd

url = "https://jobs-api14.p.rapidapi.com/v2/list"
headers = {
    "x-rapidapi-key": "e46546b67emsh1da2cc270e90508p17c823jsnc5bc19b57230",
    "x-rapidapi-host": "jobs-api14.p.rapidapi.com"
}

jobs_list = []
page = 1
max_pages = 20  
while page <= max_pages:  
    querystring = {"query": "Web Developer", "location": "United States", "page": page}
    response = requests.get(url, headers=headers, params=querystring)
    
    if response.status_code == 200:
        data = response.json()
        jobs_data = data.get("jobs", [])
        
       
        if not jobs_data:
            print("Đã tải xong tất cả dữ liệu.")
            break
        
        for job in jobs_data:
            jobs_list.append({
                "Title": job.get("title", "Not Available"),
                "Company": job.get("company", "Not Available"),
                "Location": job.get("location", "Not Available"),
                "Description": job.get("description", "Not Available"),
                "Posted Date": job.get("date", "Not Available")
            })
        
        print(f"Đã tải xong trang {page}.")
        page += 1
    else:
        print(f"Lỗi khi tải dữ liệu từ API: {response.status_code}")
        break


df = pd.DataFrame(jobs_list)

print(df)


df.to_csv("jobs_data_limited_pages.csv", index=False, encoding="utf-8")
```

```python
import re
def extract_12_digits(description):
    # Tìm kiếm 12 số ngay sau dấu $
    match = re.search(r'\$(\d{12})', description)
    
    if match:
        return match.group(1)  # Trả về 12 số sau dấu $
    return None  # Nếu không tìm thấy 12 số sau $

# Áp dụng hàm trích xuất 12 số
df['Salary'] = df['Description'].apply(extract_12_digits)
df
```

```python
import re

# Hàm để chiết xuất mức lương từ cột "Description"
def extract_salary(description):
    # Biểu thức chính quy tìm các mức lương
    match = re.search(r'\$[\d,]+(?:\s?-\s?[\d,]+|\sto\s[\d,]+)?', description)
    if match:
        return match.group(0)
    return None

# Áp dụng hàm extract_salary lên cột 'Description'
df['Salary'] = df['Description'].apply(extract_salary)

# Hàm chuyển đổi mức lương từ chuỗi sang kiểu số
def convert_salary(salary):
    if salary:
        salary = salary.replace('$', '').replace(',', '')
        if 'to' in salary:
            salary = salary.split('to')[0].strip()
        return float(salary)
    return None

# Áp dụng hàm convert_salary lên cột 'Salary'
df['Salary'] = df['Salary'].apply(convert_salary)

# Kiểm tra cột 'Salary_Raw' tồn tại và có dữ liệu hợp lệ
if 'Salary_Raw' in df.columns:
    # Lọc các hàng có mức lương thấp hơn 1000 và thay thế giá trị của 'Salary'
    df.loc[df['Salary'] < 1000, 'Salary'] = df.loc[df['Salary'] < 1000, 'Salary Range']

# Kiểm tra lại cột 'Salary' sau khi thay thế
df
```
